```python
import pandas as pd
from utilities import *
df1 = pd.read_csv("datasets/rollouts.csv")
df2 = pd.read_csv("datasets/kaggle.csv")

anl = Analizer()
# Add datasets
anl(df1, "rollouts")
anl(df2, "kaggle")
anl.draw()
```

```python
# Add analitics
class AddedAnalitics(Analitics):
	def zero(df):
		return 0

	def one(df):
		return 1

anl = Analizer(AddedAnalitics)
anl(df1, "rollouts")
anl(df2, "kaggle")
anl.draw()
```

```python
# In general any class can be used
class CustomAnalitics:
	def size(df):
		return len(df)

	def wins(df):
		return sum(df.iloc[:, -1])

	def _skipped(df):
		return 1

anl = Analizer(CustomAnalitics)
anl(df1, "rollouts")
anl(df2, "kaggle")
anl.draw()
```
